```python
# Import the libraies
import pandas as pd
import numpy as np
```

```python
# Import the data
df = pd.read_csv('agora.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Transport'].mean
```

```python
df.Transport = df['Transport'].replace(np.NaN, df['Transport'].mean())
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Encoding with Replace Function
df.Area.unique()
df.Area = df.Area.replace(['Dhaka','Ctg', 'Rangpur'],[1,2,3])
df.head()
```

```python
# Label Encoder
# Import the data
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv('agora.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df.Area = LabelEncoder().fit_transform(df.Area)
df.head()
```

```python
# One Hot Encoding
df = pd.read_csv('agora.csv')
df.head()
```

```python
df.Area.unique()
```

```python
area = pd.get_dummies(df['Area'],drop_first=True)
df = df.drop('Area',axis=1)
df.head()
```

```python
df = pd.concat([df,area],axis=1)
```

```python
df.head()
```

```python
# Ordinal Encoder
df = pd.read_csv('agora.csv')
df.head()
```

```python
df.Area.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
instance = ['Dhaka', 'Ctg', 'Rangpur']
oren = OrdinalEncoder(categories=[instance])
df.Area = oren.fit_transform(df[['Area']])
```
